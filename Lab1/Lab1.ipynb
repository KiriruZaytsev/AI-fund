# Градиентный спуск

```python
from SGD import SGD
```

## Функция Бута

```python
def booth_func(x: np.array) -> float:
    return (x[0] + 2 * x[1] - 7) ** 2 + (2 * x[0] + x[1] - 5) ** 2
```

```python
x = np.linspace(-10, 10, 500)  # Диапазон x
y = np.linspace(-10, 10, 500)  # Диапазон y
X, Y = np.meshgrid(x, y)       # Сетка

Z = np.array([[booth_func(np.array([x, y])) for x in x] for y in y])

plt.figure(figsize=(12, 10))
contour = plt.contourf(X, Y, Z, levels=100, cmap='viridis')
plt.colorbar(contour)
plt.contour(X, Y, Z, levels=10, colors='black', linewidths=0.5)

min_x, min_y = 1, 3  
plt.scatter(min_x, min_y, color='red', label='Минимум (1, 3)', s=100)
plt.annotate("Минимум (1, 3)", (min_x + 0.5, min_y), color='white', fontsize=12)

plt.title("Линии уровня функции Бута", fontsize=16)
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
optimizer = SGD(lr=0.01,
                x_bound=(-10, 10),
                y_bound=(-10, 10),
                num_iter=1000,
                seed=42)
```

```python
x_min = optimizer.optimize(booth_func)
```

```python
x_min
```

## Функция Химмельблау

```python
def himmelblau_func(x: np.array) -> float:
    return (x[0]**2 + x[1] - 11)**2 + (x[0] + x[1]**2 - 7)**2
```

```python
x = np.linspace(-5, 5, 500)  # Диапазон x
y = np.linspace(-5, 5, 500)  # Диапазон y
X, Y = np.meshgrid(x, y)       # Сетка

Z = np.array([[himmelblau_func(np.array([x, y])) for x in x] for y in y])

plt.figure(figsize=(12, 10))
contour = plt.contourf(X, Y, Z, levels=100, cmap='viridis')
plt.colorbar(contour)
plt.contour(X, Y, Z, levels=10, colors='black', linewidths=0.5)

min_x, min_y = 3, 2  
plt.scatter(min_x, min_y, color='red', label=f'Минимум ({min_x}, {min_y})', s=100)
plt.annotate("Минимум (3, 2)", (min_x + 0.5, min_y), color='white', fontsize=12)

min_x, min_y = -2.8051, 3.1313 
plt.scatter(min_x, min_y, color='red', label=f'Минимум ({min_x}, {min_y})', s=100)
plt.annotate(f'Минимум ({min_x}, {min_y})', (min_x + 0.5, min_y), color='white', fontsize=12)

min_x, min_y = -3.7793, -3.2831 
plt.scatter(min_x, min_y, color='red', label=f'Минимум ({min_x}, {min_y})', s=100)
plt.annotate(f'Минимум ({min_x}, {min_y})', (min_x + 0.5, min_y), color='white', fontsize=12)

min_x, min_y = 3.5844, -1.848 
plt.scatter(min_x, min_y, color='red', label=f'Минимум ({min_x}, {min_y})', s=100)
plt.annotate(f'Минимум ({min_x}, {min_y})', (min_x + 0.5, min_y), color='white', fontsize=12)

plt.title("Линии уровня функции Химмельблау", fontsize=16)
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
optimizer = SGD(lr=0.01,
                x_bound=(-5, 5),
                y_bound=(-5, 5),
                num_iter=1000,
                seed=1)
```

```python
x_min = optimizer.optimize(himmelblau_func)
```

```python
x_min
```
